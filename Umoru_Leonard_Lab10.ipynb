```python
ages = []
with open('students.csv', 'r') as file:
    reader = csv.reader(file)
    header = next(reader)
    age_index = header.index("Age")

    for row in reader:
        age = int(row[age_index])
        ages.append(age)

if ages:
    avg_age = sum(ages) / len(ages)
    print (f'Average age of the students is: {avg_age}')
        
```
